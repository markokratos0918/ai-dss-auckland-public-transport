# SUMO Evaluation Interpretation

Although the current implementation represents a simulated validation framework rather than a full-scale deployment, the results support the feasibility of integrating AI-driven decision support with transport simulation systems.

# Project Limitations

Several limitations were identified during the development of the prototype system:

- The Kaggle dataset represents a simulated operational environment rather than a live Auckland deployment.
- The realtime GTFS dataset currently contains a limited collection period.
- The SUMO framework currently represents a scenario preparation and validation prototype rather than a fully calibrated transport network simulation.
- Operational intervention effects are estimated rather than empirically validated using live transport control systems.
- The system currently focuses on a single-corridor validation strategy to maintain project feasibility within academic constraints.
